```python
# For wafaas csv files

# Read in the csv file.
df = pd.read_csv("~/Documents/studienprojekt/results/qextr_k001_p.csv")
# df = pd.read_csv("~/Documents/studienprojekt/results/99qt_k1_01.csv")

print(df.columns)
k_values = df['K_values']
omega_values = df['Eigenvalues']
omega_values_real = [float(i.split(" ")[0]) for i in omega_values]
print(omega_values_real)
omega_values_imag = [-float(i.split(" ")[-1][:-2]) for i in omega_values]
print(omega_values_imag)

# Plot the data
# Make dots smaller
plt.scatter(k_values, omega_values_real, label='k', s=1)
plt.show()

# Fit a curve to the data
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * x**b + c

popt, pcov = curve_fit(func, k_values, omega_values_imag)
print(popt)
print(pcov)

plt.scatter(df['K_values'], omega_values_imag, label='k', s=1)
plt.plot(k_values, func(k_values, *popt), 'r-')
plt.show()
```
